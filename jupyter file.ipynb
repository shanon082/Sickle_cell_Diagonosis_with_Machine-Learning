```python
# Step 1: Import necessary libraries
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
from tensorflow.keras import backend as k
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# Step 2: Load and preprocess the data
positive_folder = 'C:/Users/ALFRED/Desktop/shan/Positive/Unlabelled'  # Path to folder containing positive (sickle cell) images
negative_folder = 'C:/Users/ALFRED/Desktop/shan/Negative/Clear'  # Path to folder containing negative (non-sickle cell) images

from PIL import Image

# Function to load images and labels
def load_images_and_labels(folder, label):
    images = []
    labels = []
    for filename in os.listdir(folder):
        if filename.endswith(".jpg") or filename.endswith(".png"):
            img_path = os.path.join(folder, filename)
            img = Image.open(img_path)
            # Resize images to a consistent size
            img = img.resize((64, 64))
            # Convert image to array
            img_array = np.array(img)
            images.append(img_array)
            labels.append(label)
    return images, labels

# Load positive (sickle cell) images and labels
positive_images, positive_labels = load_images_and_labels(positive_folder, label=1)

# Load negative (non-sickle cell) images and labels
negative_images, negative_labels = load_images_and_labels(negative_folder, label=0)

# Combine positive and negative images and labels
images = positive_images + negative_images
labels = positive_labels + negative_labels

# Convert lists to numpy arrays
images = np.array(images)
labels = np.array(labels)

# Step 3: Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(images, labels, test_size=0.2, random_state=42)

# Step 4: Define and train the CNN model
model = Sequential([
    Conv2D(32, (3, 3), activation='relu', input_shape=(64, 64, 3)),
    MaxPooling2D((2, 2)),
    Conv2D(64, (3, 3), activation='relu'),
    MaxPooling2D((2, 2)),
    Conv2D(128, (3, 3), activation='relu'),
    MaxPooling2D((2, 2)),
    Flatten(),
    Dense(128, activation='relu'),
    Dropout(0.5),
    Dense(1, activation='sigmoid')
])

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model.summary()

# Data augmentation to prevent overfitting
datagen = ImageDataGenerator(
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest'
)

datagen.fit(X_train)

history = model.fit(datagen.flow(X_train, y_train, batch_size=32), epochs=20, validation_data=(X_test, y_test))

# Plot accuracy and loss on the same graph with different colors
plt.figure(figsize=(8, 6))

# Plot accuracy
plt.plot(history.history['accuracy'], color='blue', label='Training Accuracy')
plt.plot(history.history['val_accuracy'], color='green', label='Validation Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.title('Training and Validation Accuracy')
plt.legend(loc='upper right')

# Create a second y-axis to plot loss
plt.twinx()
# Plot loss
plt.plot(history.history['loss'], color='red', label='Training Loss')
plt.ylabel('Loss')
plt.legend(loc='lower right')

plt.show()

# Step 5: Evaluate the CNN model
cnn_predictions = model.predict(X_test)
# Convert raw predictions to class labels
cnn_predictions = (cnn_predictions > 0.5).astype(int)

cnn_accuracy = accuracy_score(y_test, cnn_predictions)
cnn_precision = precision_score(y_test, cnn_predictions)
cnn_recall = recall_score(y_test, cnn_predictions)
cnn_f1_score = f1_score(y_test, cnn_predictions)

print("CNN Model Performance:")
print("Accuracy:", cnn_accuracy)
print("Precision:", cnn_precision)
print("Recall:", cnn_recall)
print("F1 Score:", cnn_f1_score)


# Step 6: Train and evaluate other classifiers
classifiers = {
    'SVM': SVC(),
    'Random Forest': RandomForestClassifier(),
    'Logistic Regression': LogisticRegression(),
    'Decision Tree': DecisionTreeClassifier(),
    'Naive Bayes' : GaussianNB()
    
}

results = {}

for name, clf in classifiers.items():
    clf.fit(X_train.reshape(-1, 64 * 64 * 3), y_train)
    preds = clf.predict(X_test.reshape(-1, 64 * 64 * 3))
    accuracy = accuracy_score(y_test, preds)
    precision = precision_score(y_test, preds)
    recall = recall_score(y_test, preds)
    f1 = f1_score(y_test, preds)
    results[name] = {'Accuracy': accuracy, 'Precision': precision, 'Recall': recall, 'F1 Score': f1}
    
    # Print performance metrics for classifiers
    print(f"{name} Performance:")
    print("Accuracy:", accuracy)
    print("Precision:", precision)
    print("Recall:", recall)
    print("F1 Score:", f1)
    print()

# Plot performance metrics for classifiers
num_classifiers = len(results)
num_rows = (num_classifiers + 1) // 2  # Ensure at least 2 rows for better layout
num_cols = 2  # 2 columns for consistent layout

plt.figure(figsize=(14, 6 * num_rows))  # Adjust figsize based on the number of classifiers

for i, (name, metrics) in enumerate(results.items()):
    plt.subplot(num_rows, num_cols, i+1)  # Use calculated num_rows and num_cols
    plt.bar(metrics.keys(), metrics.values(), color=['blue', 'green', 'red', 'orange'])
    plt.title(name)
    plt.ylim(0, 1)
    plt.ylabel('Score')

plt.tight_layout()
plt.show()


# Step 7: Plot performance metrics for the CNN model
plt.bar(['Accuracy', 'Precision', 'Recall', 'F1 Score'], [cnn_accuracy, cnn_precision, cnn_recall, cnn_f1_score], color=['blue', 'green', 'red', 'orange'])
plt.title('CNN Model Performance')
plt.ylim(0, 1)
plt.ylabel('Score')
plt.show()


# Step 8: Plot sample images of sickle cell
plt.figure(figsize=(10, 10))
for i in range(5):
    plt.subplot(3, 3, i+1)
    plt.imshow(positive_images[i])
    plt.title("Sickle Cell")
    plt.axis('off')
plt.show()
```

```python
# Step 9: Visualize classification results

# Choose a subset of images for visualization
num_images_to_visualize = 10
visualize_images = X_test[:num_images_to_visualize]
visualize_labels = y_test[:num_images_to_visualize]

# Predict labels for the chosen images using all classifiers
classifier_predictions = {}
for name, clf in classifiers.items():
    classifier_predictions[name] = clf.predict(visualize_images.reshape(-1, 64 * 64 * 3))

# Plot the images with their predicted labels for each classifier
plt.figure(figsize=(20, 8))  # Adjust the figure size
for i in range(num_images_to_visualize):
    plt.subplot(2, 5, i + 1)
    plt.imshow(visualize_images[i])
    plt.title(f"True: {visualize_labels[i]}\n")
    plt.axis('off')
    for j, (name, preds) in enumerate(classifier_predictions.items()):
        plt.text(5, -5 - j * 10, f"{name}: {preds[i]}", fontsize=15, color='blue' if preds[i] == visualize_labels[i] else 'red')  # Adjust the vertical position and font size

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Define a function to plot images with their actual and predicted labels
def plot_images_with_labels(images, actual_labels, predicted_labels, title):
    plt.figure(figsize=(12, 6))
    for i in range(len(images)):
        plt.subplot(2, len(images)//2, i+1)
        plt.imshow(images[i])
        plt.title(f"Actual: {actual_labels[i]}\nPredicted: {predicted_labels[i]}")
        plt.axis('off')
    plt.suptitle(title)
    plt.show()

# Choose a subset of images for visualization
num_images_to_visualize = 10
visualize_images = X_test[:num_images_to_visualize]
visualize_actual_labels = y_test[:num_images_to_visualize]

# Get predicted labels for the chosen images using all classifiers
classifier_predicted_labels = {}
for name, clf in classifiers.items():
    classifier_predicted_labels[name] = clf.predict(visualize_images.reshape(-1, 64 * 64 * 3))

# Plot the images with their actual and predicted labels for each classifier
for name, predicted_labels in classifier_predicted_labels.items():
    plot_images_with_labels(visualize_images, visualize_actual_labels, predicted_labels, f"Classifier: {name}")
```
